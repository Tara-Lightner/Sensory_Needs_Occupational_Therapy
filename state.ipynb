```python
import pandas as pd
```

```python
cols = ['FIPSST','TOTKIDS_R','AutismInd_1920','sex_1920','MEDB10ScrQ5_1920','nom17_3ASD_1920','BehavSev_1920',
        'DevDelSev_1920','LearnSev_1920','AutismSev_1920','cntdiff','autism_1920','mhealth_1920','DevDelay_1920']

df = pd.read_csv('data/2019-2020 NSCH_Topical_CAHMI_DRC.csv', usecols= cols)
df.nunique()
```

```python
df["state_total"] = df.groupby('FIPSST')['FIPSST'].transform('size')
df["state_aut_total"] = df.groupby(['FIPSST','AutismInd_1920'])['AutismInd_1920'].transform('count')
df["state_prev"] = df["state_aut_total"] / df['state_total'] * 100
# Added national values manually from the internet
df['overall %'] = 2.3
df['boys %'] = 3.7
df['girls %'] = 0.9
df
```

```python
df.to_csv('data/subset_tableau.csv')
```
